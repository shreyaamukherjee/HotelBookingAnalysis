# Data Visualization

```python
cancel_per=df['is_canceled'].value_counts(normalize=True)
cancel_per
```

```python
plt.figure(figsize=(5,4))
plt.title("Cancellation Rate")
plt.bar(['Not Cancelled','Cancelled'],df['is_canceled'].value_counts())
plt.show()
df['is_canceled'].value_counts()
```

```python
plt.figure(figsize=(8,4))
plot=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
legend_labels,_=plot.get_legend_handles_labels()

plt.title("No of Hotel Reservations")
plt.xlabel("Hotels")
plt.ylabel("no of reservations")
plt.legend(['not cancelled','cancelled'])
plt.show()
```

```python
resort_hotel=df[df['hotel']=="Resort Hotel"]
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=="City Hotel"]
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(15,5))
plt.title("Daily Average Rate")
plt.plot(resort_hotel.index,resort_hotel['adr'],label='resort hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='city hotel')
plt.legend(['resort_hotel','city_hotel'])
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,6))
plot=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
legend_labels,_=plot.get_legend_handles_labels()

plt.title("Monthly Rate of Hotel Reservations")
plt.xlabel("month")
plt.ylabel("no of reservations")
plt.legend(['not cancelled','cancelled'])
plt.show()
```

```python
cancelled_data=df[df['is_canceled']==1]
top10Country=cancelled_data['country'].value_counts()[:10]
plt.title("top 10 countries got reservation cancelled")

plt.pie(top10Country, autopct="% 0.2f", labels=top10Country.index)
plt.show()
```
